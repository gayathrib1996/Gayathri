```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('Telco-Customer-Churn.xlsx')
data
```

# 1.Compare churn count wrt gender

```python
df=sns.countplot(x='gender',hue='Churn',data=data)
df
```

The distribution of churn are more or less similar for male and female.

# 2.Find out how many female senior citizens are there in the dataset.

```python
fd=data.loc[(data.SeniorCitizen==1),'gender'].value_counts()
fd
```

There are 568 female senior citizens in the dataset.

```python
fd.plot(kind='bar')
```

# 3.Compare tenure with total charges.

```python
ad=pd.read_excel("Telco-Customer-Churn.xlsx",na_values=[" "])
sns.lineplot(x='tenure',y='TotalCharges',data=ad) 
plt.title("tenure v/s total charges")
```

```python
sns.scatterplot(x='tenure',y='TotalCharges',data=ad)
plt.title("tenure v/s total charges")
```

From the above two figures we can say that as tenure increases the total charge also increase proportionately.

# 4.Find out which contract is prefered by senior citizen.

```python
dg=data.groupby('Contract')['SeniorCitizen'].count()
dg.plot(kind='bar')
plt.ylabel('No:of senior citizen')
plt.xlabel('contract')
plt.title('contract preference of senior citizen')
```

Most of the senior citizens prefers month to month contract.
