```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('StudentsPerformance.csv')
data
```

# 1. Find out how many males and females participated in the test.

```python
df=data['gender'].value_counts()
df
```

There are 518 females and 482 males who participated in the test.

# 2. What do you think about the students' parental level of education?

```python
dd=data['parental level of education'].value_counts()
dd
```

```python
dd.plot(kind='bar')
plt.title('Barchart')
plt.ylabel('frequency')
plt.xlabel('parental education level')
```

Most of the parents had gone to some college or are having some associate's degree while a few parents had completed their master's degree.


# 3. Who scores the most on average for math, reading and writing based on gender and test preparation course

```python
data[['gender','math score','writing score','reading score']].groupby(['gender']).agg('mean')
```

Female scored the most on average for writing and reading while male scored the most on average for math.

```python
data[['test preparation course','math score','writing score','reading score']].groupby(['test preparation course']).agg('mean')
```

The students who completed the test preparation course had scored the most in all the three subjects.

# 4. What do you think about the scoring variation for math, reading and writing based on gender and test preparation course.

```python
data[['gender','math score','writing score','reading score']].groupby(['gender']).agg('std')
```

The scoring variation is high among females in all the subjects.

```python
data[['test preparation course','math score','writing score','reading score']].groupby(['test preparation course']).agg('std')
```

The scoring variation is high among students those who didnt attend the test preparation course.
