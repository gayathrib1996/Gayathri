# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
df=data['Buyer Gender'].value_counts()
df
```

```python
df.plot(kind='bar')
plt.title('Barchart')
plt.ylabel('frequency')
plt.xlabel('buyer gender')
```
